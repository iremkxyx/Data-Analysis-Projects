### en çok hangi kategoride oyun var

```python
kategori_sayisi = df["Kategori"].value_counts()
print(kategori_sayisi)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(x=kategori_sayisi.index, y=kategori_sayisi.values, palette="viridis")

plt.title("Oyun Sayısına Göre Kategoriler", fontsize=16)
plt.xlabel("Kategori", fontsize=14)
plt.ylabel("Oyun Sayısı", fontsize=14)

plt.xticks(rotation=45, ha='right')

plt.show()
```

### hangi yıllarda hangi platformda oyun çıkmış

```python
pivot_table = df.pivot_table(index="Çıkış Yılı", columns='Platform', aggfunc='size', fill_value=0)
print(pivot_table)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap="YlGnBu", annot=True, fmt="d")

plt.title("Yıllara Göre Platformlarda Çıkan Oyun Sayısı", fontsize=16)
plt.xlabel("Platform", fontsize=14)
plt.ylabel("Çıkış Yılı", fontsize=14)

plt.show()
```

### kullanıcı oyu en çok olan 10 oyun

```python
top_rated_games = df.sort_values(by="Kullanıcı Oyu", ascending=False).head(10)
print(top_rated_games[["Oyun İsmi", "Kullanıcı Oyu"]])
```

```python
top_rated_games = df.sort_values(by="Kullanıcı Oyu", ascending=False).head(10)

plt.figure(figsize=(10, 6))

sns.boxplot(x="Oyun İsmi", y="Kullanıcı Oyu", data=top_rated_games, palette="Set2")

plt.title("En Yüksek 10 Kullanıcı Oyuna Sahip Oyunlar", fontsize=16)
plt.xlabel("Oyun İsimleri", fontsize=14)
plt.ylabel("Kullanıcı Oyları", fontsize=14)

plt.xticks(rotation=45, ha="right")

plt.show()
```
