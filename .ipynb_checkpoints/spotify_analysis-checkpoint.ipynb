### yıllara göre en çok dinlenen tür

```python
kategori_sayisi = df["Top Genre"].value_counts().head(10)
print(kategori_sayisi)
```

```python
sns.set(style="darkgrid")

plt.figure(figsize=(10, 6))
sns.barplot(x=kategori_sayisi.index, y=kategori_sayisi.values, palette="viridis")

plt.title("Spotify'da En Çok Dinlenen Türler", fontsize=16)
plt.xlabel("Tür", fontsize=14)
plt.ylabel("Dinlenme Sayısı", fontsize=14)

plt.xticks(rotation=45, ha='right')

plt.show()
```
